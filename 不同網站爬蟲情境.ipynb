```python
import threading
import time
def countdown(t):
    while -1 < t:
        mins, secs = divmod(t, 60)
        timer = '{:02d}:{:02d}'.format(mins, secs)
        if t > 0:
            print(timer, end="\r")
        elif t == 0:
            print(timer, end="\n")
        threading.Event().wait(1)
        #time.sleep(1)
        t -= 1
```

```python
prodlist = []
#for i  in range(1, int(jd['totalPage']) + 1):
for i  in range(1, 3):
    url = f'https://ecshweb.pchome.com.tw/search/v3.3/all/results?q=iphone&page={i}&sort=rnk/dc'
    res = requests.get(url)
    jd = json.loads(res.text)
    df0 = pd.DataFrame(jd['prods'])
    prodlist.append(df0)
    countdown(5)
df = pd.concat(prodlist, ignore_index = True)
df
```

### 抓網頁

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import datetime as dt
from dateutil.relativedelta import relativedelta

url = 'https://www.taifex.com.tw/cht/3/pcRatio'
payload = {
    'down_type': '',
    'queryStartDate': '2023/07/20',
    'queryEndDate': '2023/08/19'
}

datalist = []
base = dt.datetime.today()- relativedelta(months = 3)
while base <= dt.datetime.today():
    startdate = base.replace(day = 1).strftime('%Y/%m/%d')
    enddate = (dt.datetime(base.year, base.month+1, 1) - dt.timedelta(days = 1)).strftime('%Y/%m/%d')
    payload['queryStartDate'] = startdate
    payload['queryEndDate'] = enddate
    print(startdate, enddate)
    res = requests.get(url, params = payload)
    soup = bs(res.text, 'lxml')
    tb = soup.select('table')
    df0 = pd.read_html(tb[3].prettify())
    datalist.append(df0[0])
    base += relativedelta(months = 1)
    
df = pd.concat(datalist, ignore_index = True)
df['日期'] = pd.to_datetime(df['日期'])
df.sort_values('日期')
```

# 公開資訊觀測站
### https://mops.twse.com.tw/mops/web/t05st01

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import re
import time 

def countdown(t):
    
    while t:
        mins, secs = divmod(t, 60)
        timer = '{:02d}:{:02d}'.format(mins, secs)
        print(timer, end="\r")
        time.sleep(1)
        t -= 1
        
url = 'https://mops.twse.com.tw/mops/web/ajax_t05st01'
payload = {
    'encodeURIComponent': 1,
    'step': 1,
    'firstin': 1,
    'off': 1,
    'keyword4': '',
    'code1': '',
    'TYPEK2': '',
    'checkbtn': '',
    'queryName': 'co_id',
    'inpuType': 'co_id',
    'TYPEK': 'all',
    'co_id': '2330',
    'year': '112',
    'month': '',
    'b_date':'',
    'e_date': '',
}
payload['co_id'] = 3008 #股票代碼
payload['year'] = 111   #抓取年度
res = requests.get(url, params = payload, verify = False)
soup = bs(res.text, 'lxml')
tb = soup.select('table')
df = pd.read_html(tb[1].prettify())

new_payload = {}
for ele in soup.select('input[type=hidden]'):
    try:
        new_payload[ele['name']] = ele['value']
    except :
        continue
l=[]
url = 'https://mops.twse.com.tw/mops/web/ajax_t05st01'
for item in tb[1].select('input[type=button]'):
    print(item['onclick'])
    m = re.findall("document\.t05st01_fm\.(\w+)\.value='(\w+)'", item['onclick'])
    d = dict(m)
    new_payload.update(d)
    new_payload['encodeURIComponent'] = 1
    #print(new_payload)
    #print()
    
    while True:
        try:
            res = requests.get(url, params = new_payload, verify = False)
            res.encoding = 'utf8'
            if res.text.find('Overrun - 查詢過於頻繁,請稍後再試!!')>-1:
                raise Exception('查詢過於頻繁,請稍後再試')
                
            #print(res.text)
            #print()
            break
        except Exception as e:
            print(e)
            countdown(5)
            continue
        
    soup2 = bs(res.text, 'lxml')
    pre = soup2.select('pre')[1]
    l.append(pre.text.replace('\r\n', ''))
    
df1 = df[0]
df1['詳細資訊'] = l
df1.drop(['Unnamed: 5'], axis = 1)
df1
```

# 中時新聞網
